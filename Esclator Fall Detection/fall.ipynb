```python
from ultralytics import YOLO
import cv2
from google.colab import files

# Load yolov8 model
model = YOLO('/content/drive/My Drive/fall_detect/best_our.pt')

# Load video
cap = cv2.VideoCapture("/content/drive/My Drive/fall_detect/1.mp4")

ret = True

# Read the first frame to obtain the video dimensions
ret, frame = cap.read()
if not ret:
    print("Error: Unable to read the video frame")
    exit()

# Get the dimensions of the video frame
video_height, video_width, _ = frame.shape

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (video_width, video_height))

# Read frames
while ret:
    # Detect objects
    results = model.track(frame, persist=True)

    # Plot results
    frame_ = results[0].plot()

    # Write the processed frame to the output video file
    out.write(frame_)

    # Read the next frame
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()

# Specify the file name you want to download
file_name = "output.avi"

# Download the file
files.download(file_name)
```
